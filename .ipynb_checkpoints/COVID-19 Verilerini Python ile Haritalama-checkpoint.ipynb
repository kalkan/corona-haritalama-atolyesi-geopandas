```python
datacsv.rename(columns={'Cases[b]': 'cases', 'Deaths[c]': 'deaths', 'Location[a].1': 'location'}, inplace=True)
```

```python
datacsv.columns
```

```python
datacsv[["cases", "deaths"]] = datacsv[["cases", "deaths"]].apply(pd.to_numeric)
```

```python
datacsv.drop(columns=['Location[a]', 'Recov.[d]', 'Ref.'], inplace=True)
```

```python
datacsv
```

```python
cases = datacsv[(datacsv['cases'] > 1000)]
```

```python
cases
```

```python
cases.replace('Argentina[r]', 'Argentina', inplace=True)
cases.replace('Australia[aa]', 'Australia', inplace= True)
cases.replace('Azerbaijan[x]', 'Azerbaijan', inplace= True)
world_data.replace('Byelarus', 'Belarus', inplace= True)
cases.replace('Belgium[n]', 'Belgium', inplace = True) 
cases.replace('Brazil[f]','Brazil', inplace = True)
cases.replace('Chile[k]', 'Chile', inplace= True)
cases.replace('China[m]','China', inplace= True)
cases.replace('Cuba[ae]','Cuba', inplace= True)
cases.replace('DR Congo[ac]','Congo', inplace= True)
cases.replace('Denmark[v]','Denmark', inplace= True)
cases.replace('Egypt[p]','Egypt', inplace= True)
cases.replace('Finland[ab]','Finland', inplace= True)
cases.replace('France[l]','France', inplace= True)
cases.replace('Germany[j]','Germany', inplace= True)
cases.replace('Hong Kong','Hong Kong', inplace= True)
cases.replace('Israel[s]','Israel', inplace= True)
cases.replace('Japan[t]','Japan', inplace= True)
cases.replace('Moldova[w]','Moldova', inplace= True)
cases.replace('Morocco[y]','Morocco', inplace= True)
cases.replace('Netherlands[o]','Netherlands', inplace= True)
cases.replace('New Zealand[af]','New Zealand', inplace= True)
cases.replace('North Macedonia','Macedonia', inplace= True)
cases.replace('Norway[z]','Norway', inplace= True)
cases.replace('Russia[g]','Russia', inplace= True)
cases.replace('Serbia[u]','Serbia', inplace= True)
cases.replace('Somalia[ad]', 'Somalia', inplace= True)
cases.replace('South Sudan','Sudan', inplace= True)
cases.replace('Spain[i]','Spain', inplace= True)
cases.replace('Ukraine[q]','Ukraine', inplace= True)
cases.replace('United Kingdom[h]', 'United Kingdom', inplace= True)
cases.replace('United States[e]','United States', inplace= True)
```

```python
cases
```

```python
world_data
```

```python
cases.rename(columns = {'location' : 'CNTRY_NAME'}, inplace = True)
```

```python
cases.columns
```

```python
birlestirilmis = world_data.merge(cases, on = 'CNTRY_NAME')
```

```python
birlestirilmis
```

```python
## Gorsellestirme

ax = birlestirilmis.plot(figsize=(15,15) , column= 'cases', scheme='quantiles', legend= True, legend_kwds = {'loc': 'lower left'})
ax.set_title('COVID-19 Verileri')

import matplotlib.pyplot as plt
plt.savefig('cikti\covid_haritalama.png')
```

```python
ax2 = birlestirilmis.plot(figsize=(15,15) , column= 'cases', scheme='quantiles', k=10, legend= True, cmap= 'Reds', legend_kwds = {'loc': 'lower left'})
ax2.set_title('COVID-19 Verileri - 10 Sınıfa Ayırılmış')
```

```python
ax3 = birlestirilmis.plot(figsize=(15,15) , column= 'cases', scheme='fisher_jenks', k=10, legend= True, cmap= 'rainbow', legend_kwds = {'loc': 'lower left'})
```

```python
#vaka sayisi 100000'den fazla olanlari gorselleştirme

vaka = birlestirilmis[birlestirilmis['cases'] > 100000]
```

```python
ax4 = vaka.plot(figsize=(15,15) , column= 'cases', scheme='fisher_jenks', k=4, legend= True, cmap= 'rainbow', legend_kwds = {'loc': 'lower left'})
```
