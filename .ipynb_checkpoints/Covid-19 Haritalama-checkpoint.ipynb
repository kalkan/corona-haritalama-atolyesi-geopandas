## Kutuphaneleri Yükleme

```python
#ürettiğimiz figürlerin notebook içinde kalması için
%matplotlib inline

import pandas as pd
import geopandas as gpd
import requests
import matplotlib as plt
```

```python
r = requests.get('https://en.wikipedia.org/wiki/Template:COVID-19_pandemic_data')
data = pd.read_html(r.text)
data = data[0]
data = data[:-2] 
data.to_csv('covid.csv', index=False)
datacsv = pd.read_csv("covid.csv") 
datacsv = datacsv[1:]
datacsv.rename(columns={'Cases[b]': 'cases', 'Deaths[c]': 'deaths', 'Location[a].1': 'location'}, inplace=True)
datacsv[["cases", "deaths"]] = datacsv[["cases", "deaths"]].apply(pd.to_numeric)
datacsv.drop(columns=['Location[a]', 'Recov.[d]', 'Ref.'], inplace=True)
```

```python
# del data

cases = datacsv[(datacsv['cases'] > 1000 )]
del datacsv
# https://hub.arcgis.com/datasets/a21fdb46d23e4ef896f31475217cbb08_1?geometry=-80.859%2C-89.944%2C75.234%2C48.630
world_data = gpd.read_file(r'girdi/dunyashape.shp')
```

```python
world_data.plot()
```

```python

world_data_list = world_data['CNTRY_NAME'].tolist()
cases_list = cases['location'].tolist()


world_data_list.sort()
cases_list.sort()

for ulkeler in cases_list:
    if ulkeler in world_data_list:
        pass
    else:
        print(ulkeler + ' shape dosyasinda yok')
        
cases.to_csv('covid_v2.csv', index=False)
```

```python
cases.replace('Argentina[r]', 'Argentina', inplace= True) 
cases.replace('Australia[aa]', 'Australia', inplace= True)
cases.replace('Azerbaijan[x]', 'Azerbaijan', inplace= True)
world_data.replace('Byelarus', 'Belarus', inplace= True)
cases.replace('Belgium[n]', 'Belgium', inplace = True) 
cases.replace('Brazil[f]','Brazil', inplace = True)
cases.replace('Chile[k]', 'Chile', inplace= True)
cases.replace('China[m]','China', inplace= True)
cases.replace('Cuba[ae]','Cuba', inplace= True)
cases.replace('DR Congo[ac]','Congo', inplace= True)
cases.replace('Denmark[v]','Denmark', inplace= True)
cases.replace('Egypt[p]','Egypt', inplace= True)
cases.replace('Finland[ab]','Finland', inplace= True)
cases.replace('France[l]','France', inplace= True)
cases.replace('Germany[j]','Germany', inplace= True)
cases.replace('Hong Kong','Hong Kong', inplace= True)
cases.replace('Israel[s]','Israel', inplace= True)
cases.replace('Japan[t]','Japan', inplace= True)
cases.replace('Moldova[w]','Moldova', inplace= True)
cases.replace('Morocco[z]','Morocco', inplace= True)
cases.replace('Netherlands[o]','Netherlands', inplace= True)
cases.replace('New Zealand[af]','New Zealand', inplace= True)
cases.replace('North Macedonia','Macedonia', inplace= True)
cases.replace('Norway[y]','Norway', inplace= True)
cases.replace('Russia[g]','Russia', inplace= True)
cases.replace('Serbia[u]','Serbia', inplace= True)
cases.replace('Somalia[ad]', 'Somalia', inplace= True)
cases.replace('South Sudan','Sudan', inplace= True)
cases.replace('Spain[i]','Spain', inplace= True)
cases.replace('Ukraine[q]','Ukraine', inplace= True)
cases.replace('United Kingdom[h]', 'United Kingdom', inplace= True)
cases.replace('United States[e]','United States', inplace= True)
```

```python
cases.rename(columns = {'location': 'CNTRY_NAME'}, inplace = True)

combined = world_data.merge(cases, on = 'CNTRY_NAME')

combined.to_file(r'C:\Users\uzayuzal\Documents\GitHub\corona-haritalama-atolyesi-geopanda\cikti\covid-dunya.shp')
```

```python
combined
```

```python
import matplotlib.pyplot as plt

ax = combined.plot(figsize=(15, 15), column='cases', scheme='quantiles', legend=True, legend_kwds={'loc': 'lower left'})
ax.set_title("Coronavirus Cases Worldwide")

plt.savefig('cikti\covid-harita.png')

```

```python
ax2 = combined.plot(figsize=(15, 15), column='cases', scheme='fisher_jenks', k=10, legend=True, cmap='rainbow', edgecolor='black', legend_kwds={'loc': 'lower left'})
ax.set_title("Coronavirus Cases Worldwide (Classified with Fisher Jenks 10 Class")
```

```python
# vaka 100000'den fazla olanlar
vaka = combined[combined['cases'] > 100000]
ax3 = vaka.plot(figsize=(15, 15), column='cases', scheme='fisher_jenks', k=10, legend=True, cmap='YlGn', edgecolor='black', legend_kwds={'loc': 'lower left'})
vaka['CNTRY_NAME'].size
vaka.apply(lambda x: ax3.annotate(s=x['CNTRY_NAME'], xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
```
